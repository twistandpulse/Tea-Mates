```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
import matplotlib.gridspec as gridspec
import seaborn as sns
```

```python
df = pd.read_csv('final_model_test.csv',index_col = 0)
df.columns
```

```python
df['LogPrice']=np.log(df['SalePrice'])
```

```python
fig = plt.figure(figsize=(14, 5))
gs = gridspec.GridSpec(1, 3)
ax1 = plt.subplot(gs[0, 0])
ax2 = plt.subplot(gs[0, 1])
ax3 = plt.subplot(gs[0, 2])

df.plot.scatter(x='GrLivArea', y='LogPrice', ax=ax1)
df.plot.scatter(x='HeatingQC', y='LogPrice', ax=ax2)
df.plot.scatter(x='CentralAir', y='LogPrice', ax=ax3)
```
